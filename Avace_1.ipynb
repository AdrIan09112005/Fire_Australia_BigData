### loadDF: Caracteristicas de impresion en cada lectura de CVS

```python
def loadDF(direction): 
    # Proceso Lectura Data Frame
    df = pd.read_csv(direction)

    print("La información es: \n {}".format(df.info()))
    print("Los primeros elementos: \n {}".format(df.head()))
    print("La información estadistica: \n {}".format(df.describe()))
    return df
```

## Fire NRT M6 96619

```python
# Datos
data = {
    "latitude": [-14.281, -14.284, -14.302, -14.283, -14.285],
    "longitude": [143.636, 143.532, 143.706, 143.652, 143.521],
    "brightness": [323.9, 343.5, 320.2, 320.4, 349.4],
    "scan": [1.7, 1.7, 1.7, 1.7, 1.7],
    "track": [1.3, 1.3, 1.3, 1.3, 1.3],
    "bright_t31": [302.3, 306.3, 305.0, 303.3, 304.7],
    "frp": [26.8, 84.3, 14.1, 18.4, 110.7]
}

df = pd.DataFrame(data)

```

```python
# 1. Distribución de Brightness
plt.figure(figsize=(10, 5))
sns.histplot(df_fire_nrt_M6_96619["brightness"], bins=10, kde=True, color='b')
plt.title("Distribución de Brightness", fontsize=16)
plt.xlabel("Brightness", fontsize=12)
plt.ylabel("Frecuencia", fontsize=12)
plt.show()
```

```python
# Plot 2: Scatter plot entre Latitude y Longitude con tamaño basado en FRP
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df_fire_nrt_M6_96619["longitude"], y=df_fire_nrt_M6_96619["latitude"], size=df_fire_nrt_M6_96619["frp"], hue=df_fire_nrt_M6_96619["brightness"], palette="coolwarm", sizes=(20, 200))
plt.title("Ubicación de Puntos de Brillo con FRP", fontsize=16)
plt.xlabel("Longitude", fontsize=12)
plt.ylabel("Latitude", fontsize=12)
plt.legend(title="Brightness", loc="upper left", bbox_to_anchor=(1,1))
plt.show()
```

```python
# 3. Relación entre Brightness y Bright_t31
plt.figure(figsize=(10, 5))
sns.regplot(x=df["brightness"], y=df["bright_t31"], scatter_kws={"color": "red"}, line_kws={"color": "blue"})
plt.title("Relación entre Brightness y Bright_t31", fontsize=16)
plt.xlabel("Brightness", fontsize=12)
plt.ylabel("Bright_t31", fontsize=12)
plt.show()
```

```python
# 4. Boxplot de FRP
plt.figure(figsize=(8, 5))
sns.boxplot(y=df["frp"], color="purple")
plt.title("Distribución de FRP", fontsize=16)
plt.ylabel("FRP", fontsize=12)
plt.show()
```

```python
# 5: Heatmap de Correlación
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de Correlación entre Variables", fontsize=16)
plt.show()
```
